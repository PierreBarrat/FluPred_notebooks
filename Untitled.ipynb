```julia
param = FluTools.get_aa_parameters()
param["startdate"] = Date(2001)
sp, ph = FluTools.read_data("../data/aligned_h3n2_ha_aa.fasta", param);
```

```julia
sp_na = FluTools.StrainPop(strains = FluTools.filterstrains(sp, regions = ["north_america"]));
FluTools.bin_by_date!(sp_na, start = param["startdate"], binwidth=param["binwidth"], binspacing = param["binspacing"])
ph_na = FluTools.positions_history(sp_na);
```

```julia
all_ascending_traj_na = FluTools.get_all_trajectories(ph_na, previous_state_condition = true);
# Frequency 20% at least
minfreq = 0.2; df = (1. -minfreq)/2
trajlist_na = FluTools.frequency_condition(all_ascending_traj_na, minfreq+df, dα = df);
FluTools.get_strains!(trajlist_na, sp_na)
```

```julia
all_ascending_traj = FluTools.get_all_trajectories(ph, previous_state_condition = true);
# Frequency 20% at least
minfreq = 0.5; df = (1. -minfreq)/2
trajlist = FluTools.frequency_condition(all_ascending_traj, minfreq+df, dα = df);
FluTools.get_strains!(trajlist, sp)
```

```julia
length(trajlist)
```

```julia
lost_traj = trajlist[findall(x->x.fixation==:lost, trajlist)]
fixed_traj = trajlist[findall(x->x.fixation==:fixed, trajlist)]
histogram([length(x.t)-2 for x in fixed_traj],bins=0:1:24, label="fixed")
histogram!([length(x.t)-2 for x in lost_traj],bins=0:1:24, alpha=0.7, label="lost")

# histogram!([length(x.t)-2 for x in trajlist_na],bins=0:1:24, normalize=:probability,alpha=0.6)
```
