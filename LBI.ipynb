```julia
using MyBioTools.FluTools, TreeTools, MyBioTools
using Plots
pyplot()
using Dates
using FastaIO
include("../../make_all_plots.jl")
using Statistics
using Random
```

```julia
headers = MyBioTools.read_fasta_headers("../data/trees/tree_ha_100permonth_NEW/aligned_h3n2_ha_sub.fasta")
sp_all = FluTools.StrainPop("../data/aligned_h3n2_ha_aa.fasta");
idx = findall(x->in(x.strain, headers), sp_all.strains)
sp = FluTools.StrainPop(strains = sp_all.strains[idx]);
```

```julia
param = get_aa_parameters()
param["binwidth"] = Day(15)
param["binspacing"] = Day(30)
FluTools.bin_by_date!(sp, start = param["startdate"], binwidth=param["binwidth"], binspacing = param["binspacing"]);
```

```julia
# FluTools.get_lbi!(sp, "../data/trees/tree_ha_100permonth_NEW/tree_raw.nwk")
tree = read_tree("../data/trees/tree_ha_100permonth_NEW/tree_raw.nwk", DataType=LBIData);
```

```julia
L = length(sp.strains[1].aa_seq)
ph = Array{FluTools.PosEvo,1}(undef,L)
println("History of all positions...")
for i in 1:L
    print("$i       \r")
    ph[i] = FluTools.positions_history(i, sp)
end
```

```julia
# All trajectories
all_traj = FluTools.all_trajectories(ph)
FluTools.get_strains!(all_traj, sp)
FluTools.get_regions!(all_traj, sp)
@time FluTools.get_lbi!(sp, tree, date_leeway = Day(0));
@time FluTools.compute_fitness!(all_traj, sp, lbi=true, date=false, region=false, shift=:mean)
# Rising trajectories
ascending_traj = FluTools.previous_state_condition(all_traj, :lost);
```

## Vs frequency, 2 fitness bins

```julia
freqcond_traj_low = Dict()
freqcond_traj_high = Dict()
freqcond_traj = Dict()
alphabins = [(x,0.05) for x in 0.1:0.1:0.9]
# alphabins = [(0.3,0.05)]
for (α, dα) in alphabins
    tmp = FluTools.frequency_condition(ascending_traj, α, dα=dα)
    fitness_values = [t.fitness[t.index[:active]] for t in tmp]
    fitness_values = fitness_values[findall(!isnan, fitness_values)]
    medfit = median(fitness_values)
    freqcond_traj_low[α] = tmp[findall(x->x.fitness[x.index[:active]] < medfit, tmp)]
    freqcond_traj_high[α] = tmp[findall(x->x.fitness[x.index[:active]] >= medfit, tmp)]
    freqcond_traj[α] = tmp
end
```

```julia
dat_freqcond_traj_low = Pfix_vs_frequency(freqcond_traj_low);
dat_freqcond_traj_high = Pfix_vs_frequency(freqcond_traj_high);
dat_freqcond_traj = Pfix_vs_frequency(freqcond_traj);

p = plot(xlabel="Frequency", ylabel="Pfix", title="Ascending trajectories", size=(900,600))
plot!(p, dat_freqcond_traj_low[2], dat_freqcond_traj_low[3], err = 0.25 ./sqrt.(dat_freqcond_traj_low[4]), label="Low fitness")
plot!(p, dat_freqcond_traj_high[2], dat_freqcond_traj_high[3], err = 0.25 ./sqrt.(dat_freqcond_traj_high[4]), label="High fitness")

plot!(p, dat_freqcond_traj[2], dat_freqcond_traj[3], err = 0.25 ./sqrt.(dat_freqcond_traj[4]), label="All trajectories")
plot!(p, [0,1], [0,1], line=(1, :black, :dashdot), label="")
```

# Vary tau

```julia
function doplot(ascending_traj)
    
freqcond_traj_low = Dict()
freqcond_traj_high = Dict()
freqcond_traj = Dict()
alphabins = [(x,0.05) for x in 0.1:0.1:0.9]
# alphabins = [(0.3,0.05)]
for (α, dα) in alphabins
    tmp = FluTools.frequency_condition(ascending_traj, α, dα=dα)
    fitness_values = [t.fitness[t.index[:active]] for t in tmp]
    fitness_values = fitness_values[findall(!isnan, fitness_values)]
    medfit = median(fitness_values)
    freqcond_traj_low[α] = tmp[findall(x->x.fitness[x.index[:active]] < medfit, tmp)]
    freqcond_traj_high[α] = tmp[findall(x->x.fitness[x.index[:active]] >= medfit, tmp)]
    freqcond_traj[α] = tmp
end  
dat_freqcond_traj_low = Pfix_vs_frequency(freqcond_traj_low);
dat_freqcond_traj_high = Pfix_vs_frequency(freqcond_traj_high);
dat_freqcond_traj = Pfix_vs_frequency(freqcond_traj);

p = plot(xlabel="Frequency", ylabel="Pfix", title="Ascending trajectories", size=(900,600))
plot!(p, dat_freqcond_traj_low[2], dat_freqcond_traj_low[3], err = 0.25 ./sqrt.(dat_freqcond_traj_low[4]), label="Low fitness")
plot!(p, dat_freqcond_traj_high[2], dat_freqcond_traj_high[3], err = 0.25 ./sqrt.(dat_freqcond_traj_high[4]), label="High fitness")

plot!(p, dat_freqcond_traj[2], dat_freqcond_traj[3], err = 0.25 ./sqrt.(dat_freqcond_traj[4]), label="All trajectories")
plot!(p, [0,1], [0,1], line=(1, :black, :dashdot), label="")
    return p
end
```

```julia
tau_ref = FluTools.get_tau("h3n2","ha")*10
tauarr = [0.01,0.03,0.06,0.1,0.2] * tau_ref
plot_list = []
for τ in tauarr
    @time FluTools.get_lbi!(sp, tree, date_leeway = Day(0), τ = τ);
    @time FluTools.compute_fitness!(ascending_traj, sp, lbi=true, date=false, region=false, shift=:mean)
    push!(plot_list, doplot(ascending_traj))
end
```
